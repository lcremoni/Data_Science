# Analisi Modelli Unico

```
#@title Numero pratiche per stato
d = pd.read_excel(path+'16-09-19 Unico PF.xlsx')
d = d[['Zona','descr. Zona','Operatore ','data apertura pratica','Data Impegno','Stato Dichiarazione']]
d.columns = ['zona','descr_zona','operatore','data_apertura','data','stato']
d['descr_zona'] = d['descr_zona'].str.strip().replace({'SALO\'':'SALO','BRESCIA UNO':'BRESCIA',
                                           'BRESCIA DUE':'BRESCIA-IMPRESA VERDE BRESCIA SRL'
                                           ,'SEDE PROVINCIALE':'BRESCIA-IMPRESA VERDE BRESCIA SRL'})
d['zona']=d['descr_zona']
d.groupby('stato').count()
```

```
#@title Numero pratiche chiuse per zona
d['unico'] = d.stato.apply(lambda x: 1 if x in ['Dichiarazione chiusa','Dichiarazione stampata'] else 0)
d = d.loc[d['unico'] == 1]
d.groupby('descr_zona').count()
```

```
#@title dati statistici numero pratiche per zona
d.groupby('descr_zona')['zona'].count().describe()
```

```
#@title Grafico pratiche Coldiretti

g_data = d.groupby('data').count()

giorni = g_data.index
p = figure(title="Sistema Coldiretti", x_axis_type='datetime', plot_width=1000)
p.vbar(x=giorni, bottom=0, top=g_data['unico'], width=timedelta(days=0.8), 
       fill_alpha=1, color=Category20[4][0], line_color=None, legend="Numero Modelli Unico")
p.legend.location = "top_left"
p.xaxis[0].formatter.days = '%d/%m'
p.xaxis.major_label_orientation = 3.14/3
#p.xaxis.formatter=DatetimeTickFormatter(days = ['%d/%m'])
show(p)
```

```
g_data['unico'].describe()
```

```
d7_zonadata = d7.groupby(['zona','data']).count()
d7_zonadata['zona'] = d7_zonadata.index.get_level_values(level=0)
d7_zonadata['data'] = d7_zonadata.index.get_level_values(level=1)
d7_zonadata.data = d7_zonadata.data.astype(str)
index = pd.date_range(d7_zonadata.data.min(), d7_zonadata.data.max())
multi_index = pd.MultiIndex.from_product([d7_zonadata.zona.unique(), index])
d7_zonadata = d7_zonadata.reindex(multi_index)

d7_zonadata['data'] = d7_zonadata.index.get_level_values(level=1)
d7_zonadata['zona'] = d7_zonadata.index.get_level_values(level=0)
#d7_zonadata = d7_zonadata.fillna(0)
d7_zonadata.data = d7_zonadata.data.astype(str)
```

```
from bokeh.models import BasicTicker, ColorBar, ColumnDataSource, LinearColorMapper, PrintfTickFormatter
from bokeh.transform import transform
from bokeh.palettes import Greens

#data.Year = data.Year.astype(str)
#data = data.set_index('Year')
#data.drop('Annual', axis=1, inplace=True)
#data.columns.name = 'Month'

# reshape to 1D array or rates with a month and year for each row.
#df = pd.DataFrame(data.stack(), columns=['rate']).reset_index()

source = d7_zonadata

# this is the colormap from the original NYTimes plot
#colors = ["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce", "#ddb7b1", "#cc7878", "#933b41", "#550b1d"]
colors = list(reversed(Greens[9]))
mapper = LinearColorMapper(palette=colors, low=d7_zonadata['730'].min(), high=d7_zonadata['730'].max(),
            nan_color='gainsboro')

p = figure(plot_width=800, plot_height=350, title="Analisi 730 per zona e data",
           x_range=list(index.astype(str)), y_range=list(d7_zonadata.zona.unique()),
           toolbar_location=None, tools="", x_axis_location="above")

p.rect(x="data", y="zona", width=1, height=1, source=source,
       line_color=None, fill_color=transform('730', mapper), fill_alpha=1)

color_bar = ColorBar(color_mapper=mapper, location=(0, 0),
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%d"))

p.add_layout(color_bar, 'right')
ht = HoverTool(tooltips=[("zona", "@zona"),("data", "@data{%d/%m/%Y}"), ("730", "@730")],
          formatters={'data': 'datetime'})
p.add_tools(ht)
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "5pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = 1.0

show(p)
```

# Analisi sistema per zona

```
import numpy as np
from bokeh.palettes import Blues

d_zonadata = d.groupby(['zona','data']).count()
d_zonadata = d_zonadata['stato']
#d_zonadata.rename("unico")
ds_zonadata = ds.groupby(['zona','data']).count()
ds_zonadata = ds_zonadata['unico']
#ds_zonadata.rename("unico_sp")
dtot_zonadata = pd.merge(d_zonadata, ds_zonadata, left_index=True, right_index=True, how='outer')
d7_zonadata = d7.groupby(['zona','data']).count()
d7_zonadata = d7_zonadata['730']
dtot_zonadata = pd.merge(dtot_zonadata, d7_zonadata, left_index=True, right_index=True, how='outer')
dtot_zonadata['zona'] = dtot_zonadata.index.get_level_values(level=0)
dtot_zonadata['data'] = dtot_zonadata.index.get_level_values(level=1)
dtot_zonadata.data = dtot_zonadata.data.astype(str)
index = pd.date_range(dtot_zonadata.data.min(), dtot_zonadata.data.max())
multi_index = pd.MultiIndex.from_product([dtot_zonadata.zona.unique(), index])
dtot_zonadata = dtot_zonadata.reindex(multi_index)

dtot_zonadata['data'] = dtot_zonadata.index.get_level_values(level=1)
dtot_zonadata['zona'] = dtot_zonadata.index.get_level_values(level=0)
#dtot_zonadata = dtot_zonadata.fillna(0)
dtot_zonadata.data = dtot_zonadata.data.astype(str)
dtot_zonadata['pratiche']=dtot_zonadata[['730','unico','stato']].sum(axis=1)
dtot_zonadata['pratiche'] = dtot_zonadata['pratiche'].replace(0, np.nan)
```

```
source = dtot_zonadata

colors = list(reversed(Blues[9]))
mapper = LinearColorMapper(palette=colors, low=dtot_zonadata['pratiche'].min(), high=dtot_zonadata['pratiche'].max(),
            nan_color='gainsboro')
myrange = pd.date_range('2019-04-27', '2019-05-27').astype(str).values
p = figure(plot_width=1100, plot_height=450, title="Analisi pratiche per zona e data",
           x_range=list(index.astype(str)), y_range=list(dtot_zonadata.zona.unique()), x_axis_location="above")

p.rect(x="data", y="zona", width=1, height=1, source=source,
       line_color=None, fill_color=transform('pratiche', mapper), fill_alpha=1)

color_bar = ColorBar(color_mapper=mapper, location=(0, 0),
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%d"))

p.add_layout(color_bar, 'right')
ht = HoverTool(tooltips=[("zona", "@zona"),("data", "@data{%d/%m/%Y}"), ("pratiche", "@pratiche")],
          formatters={'data': 'datetime'})
p.add_tools(ht)
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "5pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = 1.0
show(p)
```
